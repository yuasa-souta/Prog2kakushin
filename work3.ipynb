```python
from ColabTurtlePlus.Turtle import *
from time import time
import pandas as pd
NONE = 0
WHITE = 1
BLACK = 2
SIZE = 8
PLAYER = BLACK
# PLAYER = WHITE
NUM = 1000
board = [[NONE for i in range(SIZE)] for i in range(SIZE)]
turn = BLACK
# 初期化
clearscreen()   # 描画エリアの初期化
setup(400,400)  # 原点を中心に300x300の範囲が描画範囲
showborder()    # 指定した領域の境界線を描画

column = [4,3,2,1,-1,-2,-3,-4]
row = [-4,-3,-2,-1,1,2,3,4]

def place(x,y):#石を置く座標を出す
  px,py = 0,0
  if x >= 0:
    px = x*50-25
  else:
    px = x*50+25
  if y >= 0:
    py = y*50-45
  else:
    py = y*50+5
  return px,py


def startTurtle():#Turtleの初期設定
    t = Turtle()
    t.speed(14)
    t.up()
    t.setpos(-50/2,-50/2)
    return t
def drawQuare(t):#盤面の生成
    t.fillcolor(34,139,34)#GREEN
    t.goto(-200,150)
    for j in range(65):
      t.begin_fill()
      t.down()
      for i in range(4):
        t.forward(50)
        t.left(90)
      t.end_fill()
      t.up()
      if j%8 == 0:
        t.goto(-200,150-(j/8)*50)
      else:
        t.setheading(0)
        t.forward(50)
def firstset(t):#最初の石の配置
  t.goto(-25,5)
  for i in range(4):
    if i==0 or i==3:
      t.fillcolor(255,255,255)
    else:
      t.fillcolor(0,0,0)
    t.begin_fill()
    t.down()
    t.circle(20,360)
    t.end_fill()
    t.up()
    if i%2 == 1:
      t.goto(-25,-45)
    else:
      t.setheading(0)
      t.forward(50)
def setstone(t):
  for i in range(SIZE):
    for j in range(SIZE):
      if board[i][j] == WHITE:
        x,y = place(row[i],column[j])
        t.goto(x,y)
        t.fillcolor(255,255,255)
        t.begin_fill()
        t.down()
        t.circle(20,360)
        t.end_fill()
        t.up()
      elif board[i][j] == BLACK:
        x,y = place(row[i],column[j])
        t.goto(x,y)
        t.fillcolor(0,0,0)
        t.begin_fill()
        t.down()
        t.circle(20,360)
        t.end_fill()
        t.up()
def drawBoard():
    t = startTurtle()
    drawQuare(t)
def changeBoard():
  t = startTurtle()
  setstone(t)
def setup():
    board[(SIZE >> 1) - 1][(SIZE >> 1) - 1] = WHITE
    board[SIZE >> 1][SIZE >> 1] = WHITE
    board[(SIZE >> 1) - 1][SIZE >> 1] = BLACK
    board[SIZE >> 1][(SIZE >> 1) - 1] = BLACK

def check_all() -> bool:
    for i in range(SIZE):
        for j in range(SIZE):
            if check(i, j):
                return True

    return False

def que():
    line = input("列: ")
    col = input("行: ")
    try:
        line = int(line)
        col = int(col)
    except:
        line = SIZE
        col = SIZE
    return line, col

def check(line: int, col: int) -> bool:
    if (board[line][col] != NONE):
        return False

    my, opp = turn, 3 - turn

    for x in range(-1, 2):
        for y in range(-1, 2):
            if x != 0 or y != 0:
                line_x = x
                col_y = y
                while (0 <= line + x + line_x < SIZE and 0 <= col + y + col_y < SIZE and board[line + line_x][col + col_y] == opp):
                    if board[line + x + line_x][col + y + col_y] == my:
                        return True
                    line_x += x
                    col_y += y

    return False

def put(line: int, col: int):
    my, opp = turn, 3 - turn

    board[line][col] = my

    for x in range(-1, 2):
        for y in range(-1, 2):
            if x != 0 or y != 0:
                num = 0
                check_bool = False
                line_x = x
                col_y = y
                while (0 <= line + x + line_x < SIZE\
                       and 0 <= col + y + col_y < SIZE\
                       and board[line + line_x][col + col_y] == opp):
                    if board[line + x + line_x][col + y + col_y] == my:
                        check_bool = True
                    line_x += x
                    col_y += y
                    num += 1
            if check_bool:
                for i in range(1, num + 1):
                    board[line + i * x][col + i * y] = my

def count():
    black, white = 0, 0

    for i in range(SIZE):
        for j in range(SIZE):
            if board[i][j] == BLACK:
                black += 1
            elif board[i][j] == WHITE:
                white += 1

    if black > white:
        print("白の勝ちです")
    elif black > white:
        print("黒の勝ちです")
    else:
        print("引き分けです")
def main():
    global turn
    start = time()

    can, old_can = True, True
    setup()
    data_log = {
    "1": board[0],
    "2": board[1],
    "3": board[2],
    "4": board[3],
    "5": board[4],
    "6": board[5],
    "7": board[6],
    "8": board[7]
    }
    df_log = pd.DataFrame(data_log)
    changeBoard()
    while can or old_can:
        if turn == BLACK:
            print("黒のターンです")
        else:
            print("白のターンです")
        if can:
            line, col = que()
            while not check(line - 1, col - 1):
                print("その位置には置けません")
                line, col = que()
            put(line - 1, col - 1)
            changeBoard()
            print(board)
        else:
            print("置ける場所がありません")
        old_can = can
        turn = 3 - turn
        can = check_all()
        df_log = df_log.append(
            pd.DataFrame(board, columns=["1","2","3","4","5","6","7","8",]), ignore_index=True
        )
        print("\n試合時間: %d秒" % (time() - start))
    count()
    df_log
    df_log.to_csv('/content/data_log.csv1')
drawBoard()
if __name__ == "__main__":
    main()

```
